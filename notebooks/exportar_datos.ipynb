```python
# ── Importaciones y conexión ──────────────────────────────────
import pandas as pd
import sys
sys.path.append(r'C:\customer360\src')
from config import DB_CONFIG
from sqlalchemy import create_engine

engine = create_engine(
    f"postgresql://{DB_CONFIG['user']}:{DB_CONFIG['password']}@"
    f"{DB_CONFIG['host']}:{DB_CONFIG['port']}/{DB_CONFIG['database']}"
)

print("Conexión lista")
```

```python
# ── Cargar y exportar datos a CSV ─────────────────────────────
ruta = r'C:\customer360\data'

# Dim Cliente
df_clientes = pd.read_sql("SELECT * FROM dw.dim_cliente", engine)
df_clientes.to_csv(f'{ruta}\\dim_cliente.csv', index=False)
print(f"dim_cliente.csv — {len(df_clientes):,} registros")

# Fact Transacciones
df_transacciones = pd.read_sql("SELECT * FROM dw.fact_transacciones", engine)
df_transacciones.to_csv(f'{ruta}\\fact_transacciones.csv', index=False)
print(f"fact_transacciones.csv — {len(df_transacciones):,} registros")

# Fact Prestamos
df_prestamos = pd.read_sql("SELECT * FROM dw.fact_prestamos", engine)
df_prestamos.to_csv(f'{ruta}\\fact_prestamos.csv', index=False)
print(f"fact_prestamos.csv — {len(df_prestamos):,} registros")

# Fact Pagos
df_pagos = pd.read_sql("SELECT * FROM dw.fact_pagos", engine)
df_pagos.to_csv(f'{ruta}\\fact_pagos.csv', index=False)
print(f"fact_pagos.csv — {len(df_pagos):,} registros")

# Fact Productos Contratados
df_productos = pd.read_sql("SELECT * FROM dw.fact_productos_contratados", engine)
df_productos.to_csv(f'{ruta}\\fact_productos_contratados.csv', index=False)
print(f"fact_productos_contratados.csv — {len(df_productos):,} registros")

# Dimensiones
pd.read_sql("SELECT * FROM dw.dim_fecha",    engine).to_csv(f'{ruta}\\dim_fecha.csv',    index=False)
pd.read_sql("SELECT * FROM dw.dim_canal",    engine).to_csv(f'{ruta}\\dim_canal.csv',    index=False)
pd.read_sql("SELECT * FROM dw.dim_agencia",  engine).to_csv(f'{ruta}\\dim_agencia.csv',  index=False)
pd.read_sql("SELECT * FROM dw.dim_producto", engine).to_csv(f'{ruta}\\dim_producto.csv', index=False)

print("\nTodos los CSVs exportados correctamente")
```
